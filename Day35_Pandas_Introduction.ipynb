```python
df['Revenue'] = df['amount']*df['quantity']
print(df[['order_id','amount','quantity','Revenue']].head(10))
```

```python
def classify_order(amt):
    if amt>=20000: return'High Value'
    elif amt>=5000: return'Mid Value'
    else: return'Low Value'
df['Tier'] = df['amount'].apply(classify_order)
print(df['Tier'].value_counts())
```

```python
df.rename(columns = {'amount':'Unit_Price'}, inplace = True)
print(df.columns)
df.rename(columns = {'Unit_Price':'amount'}, inplace = True)
print(df.columns)
```

```python
cat_rev = df.groupby('category')['amount'].sum().sort_values(ascending=False)
print(cat_rev)
```

```python
top5 = df.sort_values('amount', ascending = False).head(5)
print(top5[['order_id','customer','amount']])
```

```python
# Top city by revenue:
top_city = df.groupby('city')['amount'].sum().idxmax()
print('Top city:', top_city)

# Highest avg category:
top_cat = df.groupby('category')['amount'].mean().idxmax()
print('Highest avg category:' ,top_cat)

# Delivered%
pct = (df['status'] == 'Delivered').mean()*100
print(f"Delivered%: {pct:.1f}%")
```

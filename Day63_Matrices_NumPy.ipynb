### 2.1 What is a Matrix?

A **matrix** is a 2D grid of numbers — rows and columns.  
In NumPy: a 2D array with shape `(rows, cols)`.

```
sales_raw shape = (3, 4)
               ↑    ↑
          3 products  4 regions
```

**Mental model:** Think of every matrix as a table. The shape tells you `(#rows, #columns)`.

---

### 2.2 Why Matrices in DS/ML?

| You'll see this in... | Matrix represents... |
|---|---|
| Linear Regression | Feature matrix X → shape (n_samples, n_features) |
| Neural Networks | Weights W → shape (input_size, output_size) |
| PCA | Covariance matrix → shape (n_features, n_features) |
| Pandas DataFrame | `.values` → NumPy matrix internally |
| TF-IDF (NLP) | Document-term matrix → shape (n_docs, n_terms) |

---

### 2.3 The Dot Product — Core Operation

**Vector dot product** (1D × 1D → scalar):  
```
a = [1, 2, 3]  
b = [4, 5, 6]  
dot(a, b) = 1×4 + 2×5 + 3×6 = 32
```
Rule: **element-wise multiply, then sum**.

**Matrix multiplication** (2D × 1D → 1D):  
```
A shape (3,4) × b shape (4,) → result shape (3,)
Inner dimensions must match: 4 == 4 ✓
```

**The shape rule (memorise this):**  
```
(m, n) @ (n, p) → (m, p)
         ↑↑
   these must match
```

---

### 2.4 Broadcasting — NumPy's Superpower

Broadcasting lets NumPy operate on arrays of different shapes **without copying data**.

```python
# Without broadcasting (naive)
for i in range(3):
    result[i] = sales[i] * 1.18   # apply GST to each row

# With broadcasting (NumPy way)
result = sales * 1.18             # NumPy stretches 1.18 across all elements
```

**Broadcasting rules:**  
1. Shapes are compared right-to-left  
2. Dimensions must be equal OR one of them must be 1  
3. Missing dimensions are treated as 1  

```
sales_raw   shape: (3, 4)
scalar 1.18 shape: ()     → broadcast to (3, 4) ✓

sales_raw   shape: (3, 4)
prices_raw  shape: (3,)   → reshape to (3,1) first, then broadcast to (3,4) ✓
```

---

### 2.5 Common Mistakes → Fixes

| Mistake | Error | Fix |
|---|---|---|
| `(3,4) @ (3,)` | shapes not aligned | Transpose: `(3,4) @ (4,)` |
| `prices * sales` when shapes differ | operands broadcast error | `prices.reshape(3,1) * sales` |
| Treating `np.dot` and `*` the same | wrong results silently | `*` is element-wise; `@` or `np.dot` is matrix multiply |
| Forgetting `.T` for transpose | shape error | `A.T` swaps rows↔cols: `(3,4).T → (4,3)` |

**D1 (8 pts):** Simulate **one forward pass of a neural network layer**.  
A single linear layer computes: `output = X @ W + b`  
- `X` = input features, shape (5, 3) — 5 data points, 3 features each (create with `np.random.seed(42); X = np.random.randn(5,3)`)  
- `W` = weight matrix, shape (3, 2) — maps 3 inputs to 2 neurons (use `np.random.seed(7); W = np.random.randn(3,2)`)  
- `b` = bias vector, shape (2,) — `b = np.array([0.5, -0.5])`  
- Compute `output`, print its shape, and explain in a comment why the output shape makes sense.

````python
# D1: Simulate one linear layer forward pass: output = X @ W + b
# X shape (5,3) — 5 samples, 3 features
# W shape (3,2) — 3 inputs mapped to 2 neurons
# b shape (2,)  — 1 bias per output neuron, broadcast across 5 samples
# YOUR CODE HERE
print("\n=== D1 ===")
np.random.seed(42); X = np.random.randn(5, 3)
np.random.seed(7);  W = np.random.randn(3, 2)
b = np.array([0.5, -0.5])
output = X @ W + b  # (5,3)@(3,2)+(2,) → (5,2)
print(f"  Output shape: {output.shape}")
print(f"  Output (5 samples × 2 neurons):\n{np.round(output, 3)}")
````

**D2 ★ Bonus (10 pts):** Simulate **linear regression prediction** on the sales data.  
- Features `X` = first 3 columns of `sales_raw` (shape 3×3) as float  
- Target `y` = last column of `sales_raw` (West region) as float  
- Compute the **least-squares weights**: `w = (X^T X)^{-1} X^T y` using `np.linalg.inv` and `np.dot`  
- Predict `y_hat = X @ w`  
- Print actual vs predicted for all 3 products  
- Explain in a markdown cell: what does this formula actually compute?

````python
# D2★ BONUS: Least-squares linear regression using matrix math
# Formula: w = (X^T @ X)^{-1} @ X^T @ y
# This is the closed-form solution to minimising sum of squared errors
# YOUR CODE HERE
print("\n=== D2 BONUS ===")
X_lr = sales_raw[:, :3].astype(float)
y_lr = sales_raw[:, 3].astype(float)
w_lr = np.linalg.inv(X_lr.T @ X_lr) @ X_lr.T @ y_lr
y_hat = X_lr @ w_lr
for p, act, pred in zip(products, y_lr, y_hat):
    print(f"  {p}: actual={act:.0f}, predicted={pred:.1f}")
````

*D2 explanation — what does `w = (X^T X)^{-1} X^T y` actually compute?*

> This formula computes the **least‑squares weights** that minimise the sum of squared residuals (errors) between the actual target `y` and the predicted values `X @ w`.  
> Geometrically, it finds the orthogonal projection of `y` onto the column space of the feature matrix `X`. Because the sales data matrix is full‑rank (3×3 with three independent columns), the solution perfectly interpolates the three data points, resulting in zero error on the training data.

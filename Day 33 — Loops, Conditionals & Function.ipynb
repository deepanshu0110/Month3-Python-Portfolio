```python
def classify_order(amount):
    if amount >=20000: return 'High Value'
    elif amount >=5000: return 'Mid Value'
    else: return 'Low Value'
for o in orders:
    print(f"{o['order_id']} -> {classify_order(o['amount'])}")
```

```python
def status_action(status):
    if status == 'Delivered':return '✅ Ship'
    elif status == 'Returned': return '🔄 Restock'
    else: return  '❌ Cancel'
for o in orders:
    print(f"{o['order_id']} -> {status_action(o['status'])}")
```

```python
for o in orders:
    if o['category'] == 'Electronics':
        if o['amount'] >20000:
            label = 'Premium Electronics'
        else:
            label = 'Budget Electronics'
    else:
        label = 'Non-Electronics'
    print(f"{o['order_id']} -> {label}")
```

```python
def city_region(city):
    west = ['Mumbai','Ahmedabad']
    north = ['Delhi','Jaipur','Lucknow']
    south = ['Bengaluru','Hyderabad','Chennai','Kochi']
    if city in west : return 'West'
    elif city in north: return 'North'
    elif city in south: return 'South'
    else: return 'East'
for o in orders:
    print(f"{o['city']} -> {city_region(o['city'])}")
```

```python
def order_summary(order_list):
    total = len(order_list)
    delivered = sum(1 for o in order_list if o['status'] == 'Delivered')
    revenue = sum(o['amount'] for o in order_list)
    avg = revenue/total
    return{'total_orders':total,'delivered':delivered,'total_revenue':revenue,'avg_order':avg}
summary = order_summary(orders)
for k,v in summary.items(): print(f'{k}: {v}')
```

```python
def category_revenue(order_list):
    rev = {}
    for o in order_list:
        cat = o['category']
        rev[cat] = rev.get(cat, 0) + o['amount']
    return dict(sorted(rev.items(), key=lambda x: x[1], reverse=True))
for cat, r in category_revenue(orders).items():
    print(f'{cat}: ₹{r:,}')
```

```python
def full_report(order_list):
    print('=== ORDER REPORT ===')
    print(f'Total Orders: {len(order_list)}')
    status_count = {}
    for o in order_list:
        s = o['status']
        status_count[s] = status_count.get(s,0) + 1
    print('\nStatus Breakdown:')
    for s,c in status_count.items(): print(f'  {s}: {c}')
    cat_rev = {}
    for o in order_list:
        cat = o['category']
        cat_rev[cat] = cat_rev.get(cat,0) + o['amount']
    print('\nRevenue by Category:')
    for cat,rev in sorted(cat_rev.items(),key=lambda x:x[1],reverse=True):
        print(f'  {cat}: ₹{rev:,}')
    top3 = sorted(order_list,key=lambda x:x['amount'],reverse=True)[:3]
    print('\nTop 3 Orders:')
    for o in top3: print(f'  {o["order_id"]} {o["customer"]} ₹{o["amount"]:,}')
full_report(orders)
```

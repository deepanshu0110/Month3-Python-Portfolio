```python
name = "Arjun Sharma"
salary = 85000

# Line 1: Arjun Sharma earns ₹85,000 per year
# :, adds comma separator to numbers
print(f"{name} earns ₹{salary:,} per year")

# Line 2: Monthly: ₹7,083.33
# :.2f gives exactly 2 decimal places
# :, adds comma
monthly = salary / 12
print(f"Monthly: ₹{monthly:,.2f}")

# Line 3: Annual hike (10%): ₹93,500
# :.0f removes decimals (gives whole number)
hike = salary * 1.10
print(f"Annual hike (10%): ₹{hike:,.0f}")
```

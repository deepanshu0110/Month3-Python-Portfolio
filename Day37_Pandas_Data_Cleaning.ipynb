```python
import pandas as pd
import numpy as np
df = pd.read_excel("Day35_Pandas_DataCleaning.xlsx", sheet_name = 'Raw Data', header=1)
print(df.shape)
print(df.dtypes)
```

```python
# Check nulls
print(df.isnull().sum())
print(df.isnull().sum()/len(df)*100) #as%
```

```python
# Check duplicates
print(df.duplicated().sum())
print(df[df.duplicated(keep=False)]) # show all duplicate rows
```

```python
df['discount']=df['discount'].fillna(0.0)
print(df['discount'])
```

```python
df['quantity']=df['quantity'].fillna(df['quantity'].median())
print(df['quantity'])
```

```python
df = df[df['email'].str.strip() !=""]
print(df['email'])
```

```python
assert df['discount'].isnull().sum() == 0 
print(df.isnull().sum())
```

#### The discount column has the biggest percentage of missing values at 25%(5 out of 20 rows).Leaving these nulls unaddressed created a business problem because it prevents accurate financial calculations such as total revenue. Additionally missing a quarter of data for this field makes it impossible to reliably analyse the sales leading to incomplete information.Additionally there is a secondary issue i the email column, while 5% of the records are missing.While the discount nulls hinder financial analysis. the missing email represents a different data gap a missing identifier that prevents order confirmation and directly hinders customer communication.


```python
print(df[df.duplicated(subset=['customer_name','email','category','unit_price']
                    ,keep=False)])
```

```python
print(df.shape)
```

```python
df = df.drop_duplicates(subset=['customer_name','email','category','unit_price']
                    ,keep='first')
print(df)
```

```python
print(df.shape)
```

```python
df.reset_index(drop = True, inplace=True)
print(df)
```

#### Using order_id alone fails because systems often generate a unique ID for every click, meaning a duplicate transaction would still have two different IDs. In a business a duplicate order is redundent record of the same purchase identifying them prevents shipping errors and double billing the customer.

```python
df['city']=df['city'].str.strip().str.title()
print(df['city'].unique())
```

```python
df['customer_name'] = df['customer_name'].str.strip().str.title()
print(df['customer_name'].unique())
```

```python
df['valid_email'] = df['email'].str.contains('@')
invalid_count = (df['valid_email']== False).sum()
print(invalid_count)
```

```python
df['status'] = df['status'].str.strip().str.title()
print(df['status'].value_counts())
```

#### Pandas is a case sensitive so it treats 'mumbai', 'Mumbai', 'MUMBAI' as three distinct groups. For example if you use .groupby('city').sum(), instead of seeing a total for Mumbai you see three seperate rows with split totals, making your report inaccurate.

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['order_month']=df['order_date'].dt.month
print(df['order_date'])
print(df['order_month'])
```

```python
df['revenue'] = (df['quantity'] * df['unit_price']*(1-df['discount'])).round(2)
print(df['revenue'])
```

```python
df['quantity'] = df['quantity'].astype(int)
print(df['quantity'])
```

```python
assert df['order_date'].dtype == 'datetime64[ns]'
assert df['quantity'].dtype == 'int64'
assert df['discount'].dtype == 'float64'
assert df['valid_email'].dtype == 'bool'
df.info()
```

```python
assert df.isnull().sum().sum() == 0
print(df.isnull().sum())
```

```python
top_cities = df.groupby('city')['revenue'].sum().nlargest(3)
print(top_cities)
```

#### The data contains three critical issues: null discounts in 5 rows lead to incorrect revenue calculations.Action taken to apply.fillna(0) to treat missing values to zero.
#### While 4 duplicate rows consisting of 2 pairs(Arjun at rows 0 and 14, and Priya at rows 1 and 6) falsely inflate both order counts and total sales.Action take was to use .drop_duplicates() to ensure each transaction was unique.
#### Additionally inconsistent city casing(e.g.,'Mumbai' vs 'mumbai') fragments reporting by splitting single cities into multiple groups during analysis.Action taken was to standardize the column using .str.strip(),str.title() to provide a unified view of performance.
#### Addressing these errors is essential for ensuring financial accuracy and reliable business insights.
